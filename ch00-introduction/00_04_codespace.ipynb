```python
def num(num1, num2):
  
  remainder = num1% num2
  if remainder ==0:
    
    print("hmm its an even number")
  else:
    print('oh its odd')
num(2,2)  
```

```python
def largest_num(int1, int2, int3):
  return max (int1,int2, int3)
largest_num(1,2,3)
```
